# First of all : Importing Packages

```python
import numpy as np #NumPy will be used for creating numerical arrays mathematical functions 
                   #(defined as np for ease of calling)

from matplotlib import pyplot as plt # Pyplot from Matplotlib will be used for plotting graphics 
                                     #(defined as plt for ease of calling)

from mpl_toolkits.mplot3d import Axes3D  # Axes3D will be used to create three-dimensional axes for our plot
```

```python
# Data for plotting

t = np.arange(0.0, 4, 0.01)

s = 1 + np.sin(2 * np.pi * t)
z = 1 + np.cos(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, s,  marker='o', markersize=2, label='Sin')
ax.plot(t, z , label='Cos')

ax.set(xlabel='time (s)', ylabel='Voltage (mV)',title='Trigonometric functions',label="Line 1")

ax.grid(linewidth=3)

ax.legend(loc='best')

plt.xlim(0, 4.2)
plt.ylim(0, 2.4)

fig.savefig("sin-cos-low.png")

ax.grid(linewidth=0.5)

fig.savefig("sin-cos-high.png" , dpi = 300)

plt.show()
```

### Current ticks are not ideal because they do not show the interesting values (+/- 𝜋 , +/- 𝜋/2) for sine and cosine. We’ll change them such that they show only these values

```python
# Create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize=(8, 6), dpi=80)

# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)

X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

plt.xlim(X.min() * 1.2, X.max() * 1.2)
plt.ylim(C.min() * 1.2, C.max() * 1.2)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],
[r'$-1$', r'$0$', r'$+1$'])

plt.plot(X, C, color="blue", linewidth=3.5, linestyle="-")
plt.plot(X, S, color="red", linewidth=1.5, linestyle="-")

# Save figure using 200 dots per inch
plt.savefig("exercise_4.png", dpi=200)

# Show result on screen
plt.show()
```

## Annotate some points
Let’s annotate some interesting points using the annotate command. We chose the $ 2\pi / 3 $ value and we want to annotate both the sine and the cosine. We’ll first draw a marker on the curve as well as a straight dotted line. Then, we’ll use the annotate command to display some text with an arrow.

```python
# Create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize=(8, 6), dpi=80)

# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)

X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

plt.xlim(X.min() * 1.2, X.max() * 1.2)
plt.ylim(C.min() * 1.2, C.max() * 1.2)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],
[r'$-1$', r'$0$', r'$+1$'])


plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")
plt.legend(loc='upper left')


t = 2 * np.pi / 3
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 50, color='blue')
plt.annotate(r'$cos(\frac{2\pi}{3} )=-\frac{1} {2} $',
xy=(t, np.cos(t)), xycoords='data',
xytext=(-90, -50), textcoords='offset points', fontsize=16,
arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
plt.plot([t, t],[0, np.sin(t)], color='red', linewidth=2.5, linestyle="--")
plt.scatter([t, ],[np.sin(t), ], 50, color='red')
plt.annotate(r'$sin(\frac{2\pi}{3} )=\frac{\sqrt{3} }{2} $',
xy=(t, np.sin(t)), xycoords='data',
xytext=(+10, +30), textcoords='offset points', fontsize=16,
arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

# Save figure using 200 dots per inch
plt.savefig("exercise_6.png", dpi=200)

# Show result on screen
plt.show()
```

```python
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))

axes[0].hist(n)
axes[0].set_title("Default histogram")
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title("Cumulative detailed histogram")
axes[1].set_xlim((min(n), max(n)));
```

# inset

```python
fig, ax = plt.subplots()

xx = np.linspace(-0.75, 1., 100)

ax.plot(xx, xx**2, xx, xx**3)
fig.tight_layout()

# inset
inset_ax = fig.add_axes([0.3, 0.55, 0.35, 0.35]) # X, Y, width, height

inset_ax.plot(xx, xx**2, xx, xx**3)
inset_ax.set_title('zoom near origin')

# set axis range
inset_ax.set_xlim(-0.15, 0.15)
inset_ax.set_ylim(-.005, .01)

# set axis tick locations
inset_ax.set_yticks([0, 0.005, 0.01])
inset_ax.set_xticks([-0.1,0,.1]);
```
